```python
import matplotlib.pyplot as plt

models = ['ctransformers', 'Langchain', 'Lamma-cpp', 'Ollama', 'llamafile']
response_time_1_prompt = [420.82, 73.82, 30.99, 13.42, 10.71]
response_time_3_prompt = [1800.82, 1800.82, 287.25, 193.24, 38.72]

bar_width = 0.35
index = range(len(models))

plt.figure(figsize=(10, 6))

plt.bar(index, response_time_1_prompt, bar_width, label='1 Prompt')
plt.bar([i + bar_width for i in index], response_time_3_prompt, bar_width, label='3 Prompt')

plt.xlabel('Model')
plt.ylabel('Response Time (seconds)')
plt.title('Response Time Comparison for Different tools and Prompt Counts')
plt.xticks([i + bar_width / 2 for i in index], models)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['ctransformers', 'Langchain', 'Lamma-cpp', 'Ollama', 'llamafile']
response_time_1_prompt = [420.82, 73.82, 30.99, 13.42, 10.71]
response_time_3_prompt = [1800.82, 1800.82, 287.25, 193.24, 38.72]
cpu_usage_1_prompt = [15.4, 15.4, 13.2, 28.9, 16.3]
cpu_usage_3_prompt = [30.4, 30.4, 20.3, 20.9, 24.6]
memory_usage_1_prompt = [94.6, 94.6, 96.2, 99.2, 98.3]
memory_usage_3_prompt = [98.6, 98.6, 96.8, 99.2, 97.4]

# Bar colors
colors = ['#1f77b4', '#ff7f0e']

# Bar width
bar_width = 0.35
index = np.arange(len(models))

# Plot
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Response Time
axs[0].bar(index, response_time_1_prompt, bar_width, label='1 Prompt', color=colors[0])
axs[0].bar(index + bar_width, response_time_3_prompt, bar_width, label='3 Prompt', color=colors[1])
axs[0].set_xlabel('Model')
axs[0].set_ylabel('Response Time (seconds)')
axs[0].set_title('Response Time Comparison')
axs[0].set_xticks(index + bar_width / 2)
axs[0].set_xticklabels(models, rotation=45, ha='right')
axs[0].legend()
axs[0].grid(axis='y', linestyle='--', alpha=0.7)

# CPU Usage
axs[1].bar(index, cpu_usage_1_prompt, bar_width, label='1 Prompt', color=colors[0])
axs[1].bar(index + bar_width, cpu_usage_3_prompt, bar_width, label='3 Prompt', color=colors[1])
axs[1].set_xlabel('Model')
axs[1].set_ylabel('CPU Usage (%)')
axs[1].set_title('CPU Usage Comparison')
axs[1].set_xticks(index + bar_width / 2)
axs[1].set_xticklabels(models, rotation=45, ha='right')
axs[1].legend()
axs[1].grid(axis='y', linestyle='--', alpha=0.7)

# Memory Usage
axs[2].bar(index, memory_usage_1_prompt, bar_width, label='1 Prompt', color=colors[0])
axs[2].bar(index + bar_width, memory_usage_3_prompt, bar_width, label='3 Prompt', color=colors[1])
axs[2].set_xlabel('Model')
axs[2].set_ylabel('Memory Usage (%)')
axs[2].set_title('Memory Usage Comparison')
axs[2].set_xticks(index + bar_width / 2)
axs[2].set_xticklabels(models, rotation=45, ha='right')
axs[2].legend()
axs[2].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
